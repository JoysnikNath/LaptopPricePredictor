# **Dropping or Removing Unecessary Columns**

```python
ld = ld.drop("Unnamed: 0",axis=1)
```

```python
ld = ld.join(pd.get_dummies(ld.Company))
```

```python
ld = ld.join(pd.get_dummies(ld.TypeName))
```

```python
ld = ld.drop("Company", axis = 1)
```

```python
ld = ld.drop("TypeName",axis = 1)
```

```python
ld.head()
```

```python
ld.dtypes
```

```python
ld["Ram"] = ld.Ram.str.split("GB").apply(lambda x: x[0])
```

```python
ld["Weight"] = ld.Weight.str.split("kg").apply(lambda x: x[0])
```

```python
ld['Ram'] = ld['Ram'].astype(int)
```

```python
ld['Weight'] = ld['Weight'].astype(float)
```

```python
ld.dtypes
```

```python
ld
```

```python
ld["MemoryType"] = ld.Memory.str.split().apply(lambda x: x[1])
```

```python
ld["MemorySpace"] = ld.Memory.str.split().apply(lambda x: x[0])
```

```python
def turn_mem_to_MB(value):
  if("GB" in value):
    return float(value[:value.find("GB")])*1024
  elif("TB" in value):
    return float(value[:value.find("TB")])*1048576
```

```python
ld["MemorySpace"] = ld.MemorySpace.apply(turn_mem_to_MB)
```

```python
ld = ld.drop("Memory", axis = 1)
```
