# **Splitting Memory Type and Memory Space**

```python
ld["MemoryType"] = ld.Memory.str.split().apply(lambda x: x[1])
```

```python
ld["MemorySpace"] = ld.Memory.str.split().apply(lambda x: x[0])
```

## **Conversion Function for GB/TB to MB**

```python
def turn_mem_to_MB(value):
  if("GB" in value):
    return float(value[:value.find("GB")])*1024
  elif("TB" in value):
    return float(value[:value.find("TB")])*1048576
```

## **Applying the Function to MemorySpace Table**

```python
ld["MemorySpace"] = ld.MemorySpace.apply(turn_mem_to_MB)
```
